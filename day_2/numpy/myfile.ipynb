### 1. Create `x` and compute `y = np.where(x<0, x**2, np.sqrt(x+1))`. Plot with axis labels and grid.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([-2, -1, 0, 1, 2, 3, 4, 5])
y = np.where(x<0, x**2, np.sqrt(x+1))

%matplotlib inline

plt.plot(x,y,'*',lw = 1)
plt.title("y(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
