3. Simulate a random walk: `steps = np.random.choice([-1,1], size=2000)`, `pos = steps.cumsum()`. Plot and annotate the maximum distance from start.

```python
steps = np.random.choice([-1,1], size=2000)
pos = steps.cumsum()

max_step = np.argmax(np.abs(pos))

plt.figure()
plt.plot(pos)
plt.title("Random Walk")
plt.xlabel("Step")
plt.ylabel("Position")
plt.grid()
plt.annotate('Distance to BIOSINF', xy=(max_step,pos[max_step]), xytext=(max_step,pos[max_step]),arrowprops=dict(facecolor='black', shrink=0.05),)
plt.show()

print(max_step)

```

### 4. Make a 6×6 array. Set a checkerboard pattern to −1 using slicing. Plot before and after.


```python
x = np.zeros((6, 6), dtype=int)
plt.plot(x)
plt.show()
x[1::2, ::2] = -1
x[::2, 1::2] = -1
print(x)
plt.plot(x)
plt.show()
```

```python

def printcheckerboard(n):
    print("Before")

    # create a n * n matrix
    x = np.zeros((n, n), dtype=int)
    plt.plot(x)
    plt.show()
    # fill with 1 the alternate cells in rows and columns
    x[1::2, ::2] = -1
    x[::2, 1::2] = -1

    # print the pattern
    for i in range(n):
        for j in range(n):
            print(x[i][j], end=" ")
        print()
    print("After")
    plt.plot(x)
    plt.show()

# driver code
n = 6
printcheckerboard(n)
```

## B) Images


```python
import numpy as np
import matplotlib.pyplot as plt

# Quality-of-life: render plots inline and increase figure DPI
%matplotlib inline
plt.rcParams['figure.dpi'] = 120

from PIL import Image
from IPython.display import Audio, display

import os
```

1. Load and show an image. Compute per-channel means.

```python
img_path = "dog.png"

if img_path and os.path.exists(img_path):
    im = Image.open(img_path).convert("RGB")
else:
    w, h = 150, 100
    x = np.linspace(0, 1, w)
    y = np.linspace(0, 1, h)
    xv, yv = np.meshgrid(x, y)
    arr = np.stack([xv, yv, 0.5*np.ones_like(xv)], axis=-1)
    im = Image.fromarray(np.uint8(arr*255), mode="RGB")

plt.figure()
plt.imshow(im)
plt.title("Image Display (RGB)")
plt.axis("off")
plt.show()
```

```python
array = np.array(im)
channels_mean = array.mean(axis=(0, 1))
print("Per-channel means:", channels_mean)
```

2. Convert to grayscale with `gray = img.mean(axis=2).astype(img.dtype)`. Display it.

```python
gray = array.mean(axis=2).astype(array.dtype)
plt.figure()
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

3. Swap red and blue channels using indexing.

```python
r = array[:, :, 0]
b = array[:, :, 2]
swapped = array.copy()
swapped[:, :, 0] = b
swapped[:, :, 2] = r
plt.figure()
plt.imshow(swapped)
plt.title("Red and Blue Channels Swapped")
plt.axis("off")
plt.show()
```

4. Multiply the image by a vertical gradient tint using `np.linspace(0.7, 1.3, H)[:,None,None]`.

```python
H = array.shape[0]
gradient = np.linspace(0.7, 1.3, H)[:, None, None]
gradient=array * gradient
gradient = gradient/np.max(gradient) * 255
gradient = gradient.astype(array.dtype)
plt.figure()
plt.imshow(gradient)
plt.title("Image with Vertical Gradient")
plt.axis("off")
plt.show()
```

## C) Audio


 Load and normalize an audio file. Plot 0.05 s.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile

audio_path = 'iphone_ringtone.wav'
sr = 16000
sr, data = wavfile.read('iphone_ringtone.wav')

if data.dtype == np.int16:
    norm_data = data / np.max(np.abs(data))
elif data.dtype == np.float32 or data.dtype == np.float64:
    norm_data = data / np.max(np.abs(data))
else:
    raise ValueError(f"Unsupported audio data type: {data.dtype}")

wavfile.write('iphone_ringtone_normalized.wav', sr, (norm_data * 32767).astype(np.int16))

duration = 0.05
num_samples = int(sr * duration)
plt.figure(figsize=(10, 4))
plt.plot(np.arange(num_samples) / sr, norm_data[:num_samples])
plt.tight_layout()
plt.show()
```

Concatenate the signal to itself with `np.concatenate`.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile

audio_path = 'iphone_ringtone.wav'
sr = 16000
sr, data = wavfile.read('iphone_ringtone.wav')

if data.dtype == np.int16:
    orig_data_float = data / 32768.0
else:
    orig_data_float = data.astype(np.float32)

concat_data = np.concatenate((orig_data_float, norm_data))
wavfile.write('iphone_ringtone_concat.wav', sr, (concat_data * 32767).astype(np.int16))
```

Slice out a middle 1-second clip using sample indices.

```python
if data.dtype == np.int16:
    data = data.astype(np.float32)

middle = len(data) // 2


data[middle - 8000:middle + 8000 + 1] *= 0.5
wavfile.write('iphone_ringtone_middle.wav', sr, (data * 32767).astype(np.int16))
```

Apply different gains to the first half (0.8) and second half (1.2) using a mask.

```python
half=len(data) // 2

data[:half] *= 0.8
wavfile.write('iphone_ringtone_first_half.wav', sr, (data * 32767).astype(np.int16))
data[half:] *= 1.2
wavfile.write('iphone_ringtone_second_half.wav', sr, (data * 32767).astype(np.int16))
```
